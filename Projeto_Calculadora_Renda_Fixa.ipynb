# Construção de uma calculadora de renda fixa em Python - NTN-F

```python
import datetime 
from dateutil.relativedelta import relativedelta
import urllib
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
```

Foi criada uma classe CalculadoraRendaFixa com as funções exigidas pelo trabalho.

Os objetos dessa classe podem ser usados de duas formas:
- 1) Utilizar o método calcula_titulo
- 2) Utilizar os métodos independentes

O método calcula_titulo, além de calcular o preço unitário do título, guarda o cashflow no dicionário "titulos", que é um atributo da calculadora.  

Os métodos independentes retornam o valor calculado.

**Observação: alguns métodos manipulam informações que são utilizadas no cashflow. Essas informações são armazenadas no dicionário atributo "titulos". Uma nova chave desse dicionário só é criada ao utilizar o método "calcula_titulo", portanto se não utilizar o método "calcula_titulo", não haverá uma chave associada às informações. Então existe tratamento de erro quando se utiliza os outros métodos para evitar armazenar informações no dicionário**

```python
class CalculadoraRendaFixa():

    
    def __init__(self):
        ''' Cria a calculadora com a lista de feriados obtida na tabela da ANBIMA''' 
        # Tenta obter os feriados. Caso o endereço esteja incorreto, a lista de feriados recebe None
        try:
            feriados = pd.read_excel('https://www.anbima.com.br/feriados/arqs/feriados_nacionais.xls')['Data']
            # Guarda apenas as datas dos feriados
            self.feriados = feriados[feriados.apply(lambda x: isinstance(x,datetime.datetime))].apply(lambda x: self.padroniza_data(x)).values.astype('datetime64[D]')
        
        except urllib.error.HTTPError as exception:
            print('Não foi possível obter os feriados, utilize o método obter_feriados com o endereço correto da planilha da ANBIMA antes de calcular algum título')
            self.feriados = None
        
        # Variáveis auxiliares para criação do DataFrame com cashflow de cada título
        self.titulo_atual = None
        self.data_atual = None
        
        # Dicionário com cashflow dos títulos já calculados
        self.titulos = {}

        
    def obter_feriados(self,endereco):
        ''' Atualiza a lista de feriados
            endereco: endereco do arquivo .xls da ANBIMA com a lista de feriados'''
        #Código parecido com a função de inicializar a calculadora
        try:
            feriados = pd.read_excel(endereco)['Data']
            self.feriados = feriados[feriados.apply(lambda x: isinstance(x,datetime.datetime))].apply(lambda x: self.padroniza_data(x)).values.astype('datetime64[D]')
            
        except urllib.error.HTTPError as exception:
            print('Não foi possível atualizar a lista de feriados, verifique o endereço inserido') 
            
            
    def padroniza_data(self,data, formato='%d/%m/%Y'):
        ''' Transforma a entrada para o tipo datetime.date
        data: pode ser do tipo datetime.date, datetime.datetime, str
        formato: se a data for do tipo "str", o formato padrão é "dia/mês/ano"  ''' 
        if isinstance(data,datetime.datetime):
            return data.date()
        
        if isinstance(data,datetime.date):
            return data
        
        if type(data) == str:
            try:
                return datetime.datetime.strptime(data, formato).date()       
            except ValueError:
                print('O formato padrão da data é "d/m/Y", utilize a variável "formato" para definir outra entrada')
                return None   

            
    def conta_dias(self,data_inicial,data_final):
        '''Conta dias corridos'''
        DC = (data_final - data_inicial).days
        try:
            self.titulos[self.titulo_atual].loc[self.data_atual,'DC'] += DC #guarda a quantidade de dias no DataFrame cashflow
        except KeyError:
            return DC
        return DC

    
    def conta_dias_uteis(self,data_inicial,data_final):
        '''Conta dias uteis'''
        DU = np.busday_count(data_inicial,data_final, holidays=self.feriados)-1
        try:
            self.titulos[self.titulo_atual].loc[self.data_atual,'DU'] += DU #guarda a quantidade de dias no DataFrame cashflow
        except KeyError:
            return DU
        return DU
    
    
    def calcula_prazo(self, dt_ini, dt_fim, convencao):
        '''Calcula prazo de acordo com a convencao'''
        if convencao == 'DU/252':
            return self.conta_dias_uteis(dt_ini,dt_fim)/252
        
        if convencao == 'DC/360':
            return self.conta_dias(dt_ini,dt_fim)/360

        
    def constroi_fluxo(self, dt_ini, dt_fim, frequencia, corrige_n_uteis='forward'):
        '''Constroi sequência de datas espaçadas com a frequência pré determinada
        corrige_n_uteis: corrige o fluxo para dias não uteis, se for "backward" corrige para o primeiro dia útil anterior ao não útil, se for "forward" corrige para o primeiro dia útil posterior ao não útil
        '''
        dias = 0
        semanas = 0
        meses = 0
        anos = 0

        if frequencia == 'dia':
            dias = 1
        elif frequencia == 'semana':
            semanas = 1
        elif frequencia == 'mes':
            meses = 1
        elif frequencia == 'bimestre':
            meses = 2
        elif frequencia == 'trimestre':
            meses = 3
        elif frequencia == 'semestre':
            meses = 6
        elif frequencia == 'ano':
            anos = 1

        # Cria a lista de datas começando da data final até a data inicial
        fluxos = []
        while dt_fim > dt_ini:
            if dt_fim > dt_ini:
                fluxos.append(dt_fim) # adiciona a data na lista
            dt_fim -= relativedelta(days=dias, weeks=semanas, months=meses, years=anos) # calcula próxima data do fluxo baseado na frequência

        fluxos.sort() #organiza em ordem crescente
        fluxos = np.busday_offset(fluxos, offsets=0, holidays=self.feriados, roll=corrige_n_uteis).astype(datetime.date) # corrige dias não uteis
        return fluxos

    
    def calcula_pu(self,VF, prazo_anual, taxa_anual):
        '''Traz a valor presente o valor futuro de um fluxo'''
        fator_desconto = (1+taxa_anual)**prazo_anual
        try:
            self.titulos[self.titulo_atual].loc[self.data_atual,'VF'] += round(VF,4) # guarda o VF no DataFrame cashflow
            self.titulos[self.titulo_atual].loc[self.data_atual,'Fator desconto'] += fator_desconto # guarda o fator de desconto no DataFrame cashflow
        except KeyError:
            pass
        
        return VF/fator_desconto

    
    def calcula_taxa_anual(self,PU, prazo_anual, valor_base=100):
        return (valor_base/PU)**(1/prazo_anual) - 1 
    
    
    def calcula_pu_ntnf(self, dt_base, dt_venc, tir):
        '''Calcula o preço unitário de um NTN-F
        dt_base: data de precificação do titulo
        dt_venc: data de vencimento
        tir: taxa de retorno interna (o equivalente que o título paga ao ano)'''
        
        VN = 1000 #valor nominal do título no vencimento
        taxa_cupom = 0.1 #taxa do cupom ao ano
        
        dt_base = self.padroniza_data(dt_base)
        dt_venc = self.padroniza_data(dt_venc)
        
        datas_cupons = self.constroi_fluxo(dt_base, dt_venc, 'semestre', 'forward') #constroi datas do fluxo de caixa
        cupom_sem = ((1 + taxa_cupom)**(1/2) - 1)*VN #calcula o valor do cupom
        
        PU = 0
        for data in datas_cupons:  
            #atualiza a variável auxiliar com a data para que outras funções possam guardar valores no DataFrame de cashflow
            self.data_atual = data 
            
            try:
                self.titulos[self.titulo_atual].loc[self.data_atual] = [0, 0, 0, 0] #cria uma nova linha no DataFrame com zeros
            except KeyError:
                pass
            
            prazo_anualizado = self.calcula_prazo(dt_base, data, 'DU/252')
            pu = self.calcula_pu(cupom_sem, prazo_anualizado, tir) #traz a valor presente os cupons
            
            try:
                self.titulos[self.titulo_atual].loc[self.data_atual,'PU'] += round(pu,4)
            except KeyError:
                pass
            
            PU += pu
        
        
        
        try:
            self.titulos[self.titulo_atual].loc[self.data_atual,'DU'] = 0 #corrige dias uteis para o ultimo fluxo de caixa
        except KeyError:
            pass
        pu = self.calcula_pu(VN, self.calcula_prazo(dt_base, data, 'DU/252'), tir) #traz a valor presente o valor nominal
        
        
        try:
            self.titulos[self.titulo_atual].loc[self.data_atual,'PU'] += round(pu,4)
        except KeyError:
            pass            
        PU += pu
        
        try:
            self.titulos[self.titulo_atual].loc[dt_base, 'VF'] = round(-PU,4) #Adiciona o preço unitário calculado na lista de VF no DataFrame de cashflow
            self.titulos[self.titulo_atual].sort_index(inplace=True) #ordena o DataFrame de cashflow pela data
        except KeyError:
            pass
        
        return PU

    
    def calcula_titulo(self, nome, tipo, dt_base, dt_venc, tir):
        '''Calcula o preço de um título definido pelo tipo
        nome: nome do título que será calculado
        tipo: tipo de investimento em renda fixa (por enquanto apenas NTN-F)
        dt_base: data de precificação do título
        dt_venc: data de vencimento
        tir: taxa interna de retorno'''        
        self.titulo_atual = nome #variável auxiliar para armazenar o cashflow do título calculado
        
        if tipo == 'NTNF':
            self.titulos[nome] = pd.DataFrame(columns=['VF','DU', 'Fator desconto', 'PU'])
            self.calcula_pu_ntnf(dt_base, dt_venc, tir)
            
        self.titulo_atual = None
        
    def plot_fluxo(self,nome_titulo):
        try:
            plt.figure(figsize=[15,10])
            markerline, stemlines, _ = plt.stem(self.titulos[nome_titulo].index, self.titulos[nome_titulo]['VF'], linefmt='k', markerfmt="None", basefmt='k')
            plt.setp(stemlines, 'linewidth', 8)
            plt.xticks(self.titulos[nome_titulo].index, fontsize=12, weight = 'bold', rotation=-60)
            plt.yticks(fontsize=12,weight = 'bold')
            return plt.show()
        except KeyError:
            print(f'O título {nome_titulo} não foi calculado')
            return None
```
